# Examine the temporal dynamics over the training

```python
import meta, examine
```

## ADAM, Zero pretrain

Epoch of interests:
- 1: start
- 10: 50% PHO, 0 SEM
- 40: 90% PHO, 50% SEM
- 100: Somewhat asymptotic performance
- 300: Overtrain

```python
cfg_json = 'models/fast_no_pretrain/model_config.json'
cfg = meta.Config.from_json(cfg_json)
exam = examine.Examine(cfg)
```

### Activation == 1

```python
sel_epochs = [1, 10, 40, 100, 300]
[exam.plot_input(task='triangle', epoch=x, act=1, save=False, ylim=(-5,20)) for x in sel_epochs]
```

### Activation == 0

```python
sel_epochs = [1, 10, 40, 100, 300]
[exam.plot_input(task='triangle', epoch=x, act=0, save=False, ylim=(-20,5)) for x in sel_epochs]
```

```python
exam.mean_bias
```

- With no pretraining, the model works similar to a OP model, with the main driving force being the direct connection from orthography. 
- As the model learns, the input strength continue to increase. 
- Given the overall performance in SGD is somewhat similar to ADAM, perhaps they have similar temporal dynamics? Let's check...
